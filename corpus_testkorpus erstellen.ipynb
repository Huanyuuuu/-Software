```python
#1750-1800

import os
import shutil


source_folder = "/Users/dou/Desktop/25SS/hausarbeit/semantic_change/element_corpus/1750-1800"
# Neuer Ordnerpfad
target_folder = "/Users/dou/Desktop/25SS/hausarbeit/semantic_change/corpus/1750-1800"

# Zielordner erstellen
os.makedirs(target_folder, exist_ok=True)

folder_counters = {}


for root, dirs, files in os.walk(source_folder):
    for file in files:
        if file.endswith(".txt"):
            source_file = os.path.join(root, file)

            # Unterordnernamen abrufen
            rel_path = os.path.relpath(root, source_folder)  
            folder_name = rel_path.replace(os.sep, "_")  # Mehrstufige Ordner werden durch Unterstriche getrennt.

            # Initialisiere den Zähler
            if folder_name not in folder_counters:
                folder_counters[folder_name] = 1

            # neuen Dateinamen erstellen
            new_filename = f"{folder_name}_{folder_counters[folder_name]}.txt"
            folder_counters[folder_name] += 1

            # In neuem Pfad speichern
            target_file = os.path.join(target_folder, new_filename)
            shutil.copy2(source_file, target_file)


print("fertig！")
```

```python
#Vorverarbeitung

import os
import re

def clean_text(text):
    # OCR-Wortsegmentierung korrigiert 
    text = re.sub(r"-\s*\n\s*", "", text)

    #Absätze teilen
    paragraphs = re.split(r"\n\s*\n", text)

    cleaned_paragraphs = []
    for para in paragraphs:
        # Zeilenumbrüche innerhalb von Absätzen entfernen
        para = para.replace("\n", " ")

        #In Kleinbuchstaben umwandeln
        para = para.lower()

        # Überflüssige Leerzeichen
        para = re.sub(r"\s+", " ", para).strip()

        if para:
            cleaned_paragraphs.append(para)

    #Absätze mit Leerzeilen wieder verbinden
    return "\n\n".join(cleaned_paragraphs)


# Durchsuche alle .txt-Dateien im Zielordner und cleaning
for file in os.listdir(target_folder):
    if file.endswith(".txt"):
        file_path = os.path.join(target_folder, file)

        with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
            raw_text = f.read()

        cleaned_text = clean_text(raw_text)

        #  Überschreiben Speichern
        with open(file_path, "w", encoding="utf-8") as f:
            f.write(cleaned_text)



print(" Alle Dokumente wurden vorverarbeitet!")

```

```python
#Kleine Test Korpus
#Fünf pro Jahr

import os
import re
import random
import shutil
from collections import defaultdict

# Quellordner
BASE_DIR = "/Users/dou/Desktop/25SS/hausarbeit/semantic_change/corpus/1750-1800"

# Zielordner）
OUTPUT_DIR = "/Users/dou/Desktop/25SS/hausarbeit/semantic_change/corpus_test/t1"

# Ausgabeverzeichnis erstellen
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Jahr extrahieren
year_pattern = re.compile(r'_(\d{4})_')

# Nach Jahr gruppiert
files_by_year = defaultdict(list)

for fname in os.listdir(BASE_DIR):
    if not fname.endswith(".txt"):
        continue
    match = year_pattern.search(fname)
    if match:
        year = match.group(1)
        files_by_year[year].append(os.path.join(BASE_DIR, fname))

# Jedes Jahr werden fünf zufällig ausgewählt und in dasselbe Verzeichnis kopiert.
for year, files in files_by_year.items():
    if len(files) <= 5:
        sampled = files  # Wenn es weniger als fünf sind, nimm alle.
    else:
        sampled = random.sample(files, 5)

    #Kopieren  direkt in OUTPUT_DIR, ohne für jedes Jahr Unterordner zu erstellen.
    for f in sampled:
        shutil.copy(f, OUTPUT_DIR)

print("fertig！")

```
